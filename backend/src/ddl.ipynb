```python
import sqlite3
```

```python
conn = sqlite3.connect("database.db")
```

```python
cursor = conn.cursor()
cursor.execute('select * from user')
cursor.fetchall()
```

```python
# drop all non-system tables in the connected database
cursor = conn.cursor()

cursor.execute("SELECT name FROM sqlite_master WHERE type='table' AND name NOT LIKE 'sqlite_%';")
tbls = [row[0] for row in cursor.fetchall()]
if not tbls:
    print("No tables to drop.")
else:
    for t in tbls:
        cursor.execute(f'DROP TABLE IF EXISTS "{t}";')
    conn.commit()
    print("Dropped tables:", tbls)
```

```python
user = """
create table if not exists user (
    userid INTEGER PRIMARY KEY,
    email TEXT,
    password TEXT,
    name TEXT
);
"""

requested_event = """
create table if not exists requested_event (
    eventid INTEGER PRIMARY KEY,
    userid_tutee INTEGER,
    available_start_time DATETIME,
    available_end_time DATETIME,
    category TEXT,
    title TEXT,
    description TEXT,
    uid_tutor INTEGER,
    possible_tutors TEXT,
    is_accepted BOOLEAN,
    is_deleted BOOLEAN
);
"""

meeting = """
create table if not exists meeting (
    eventid INTEGER,
    start_time DATETIME,
    end_time DATETIME
);
"""

ddl = [user, requested_event, meeting]
```

```python
for x in ddl:
    cursor = conn.cursor()
    cursor.execute(x)
    conn.commit()
```

```python
cursor.execute("SELECT name, sql FROM sqlite_master WHERE type='table' ORDER BY name;")
tables = cursor.fetchall()
if not tables:
    print("No tables found.")
else:
    for name, sql in tables:
        print(f"Table: {name}")
        print("-" * 40)
```
